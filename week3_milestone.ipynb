```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CLEANED dataset you just created
file_path = r'D:\RoadSafe_Analytics\data\US_Accidents_Cleaned_Final.csv'
df = pd.read_csv(file_path)

# Set the visual style
sns.set_theme(style="darkgrid")
print("✔ Cleaned data loaded for Week 3 Analysis.")
```

```python
# 1. Bar Graph for Severity
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Severity', hue='Severity', palette='viridis', legend=False)
plt.title('Distribution of Accident Severity (1=Low, 4=High)', fontsize=14)
plt.show()

# 2. Pie Chart for Percentage Breakdown
severity_counts = df['Severity'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Percentage Breakdown of Accident Severity')
plt.show()
```

```python
# A. Accidents by Hour of Day (Histogram)
plt.figure(figsize=(12, 6))
sns.histplot(df['Hour'], bins=24, kde=True, color='teal')
plt.title('Accident Frequency by Hour of Day', fontsize=14)
plt.xlabel('Hour (0-23)')
plt.xticks(range(0, 24))
plt.show()

# B. Accidents by Day of the Week (Bar Graph)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Weekday', order=day_order, hue='Weekday', palette='magma', legend=False)
plt.title('Accident Frequency by Day of the Week', fontsize=14)
plt.show()

# C. Accidents by Month (Bar Graph)
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='Month', order=month_order, hue='Month', palette='rocket', legend=False)
plt.xticks(rotation=45)
plt.title('Accident Frequency by Month', fontsize=14)
plt.show()
```

```python
# 1. Top 10 Weather Conditions
plt.figure(figsize=(12, 6))
top_weather = df['Weather_Condition'].value_counts().head(10)
sns.barplot(x=top_weather.values, y=top_weather.index, hue=top_weather.index, palette='Blues_d', legend=False)
plt.title('Top 10 Weather Conditions during Accidents', fontsize=14)
plt.xlabel('Number of Accidents')
plt.show()

# 2. Most Common Road Types/Infrastructure
road_features = ['Junction', 'Traffic_Signal', 'Crossing', 'Station', 'Stop']
feature_counts = [df[f].sum() for f in road_features]

plt.figure(figsize=(10, 5))
sns.barplot(x=road_features, y=feature_counts, hue=road_features, palette='coolwarm', legend=False)
plt.title('Accidents by Road Infrastructure Type', fontsize=14)
plt.ylabel('Total Accidents')
plt.show()
```
